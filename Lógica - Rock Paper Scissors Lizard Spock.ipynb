```python
#PEDRA PAPEL OU TESOURA

import random

rps_options = ['pedra', 'papel', 'tesoura']

def choicesPedraPapelTesoura(choice_m,choice_u):
    print("Usuário escolheu", choice_u) 
    print("Máquina escolheu", choice_m) 


#return 0 empate, return 1 vitória player 1, return 2 vitória player 2
def pedraPapelTesoura(i_player1,i_player2):
    
    i_player1 = i_player1.upper()
    i_player2 = i_player2.upper()
    
    if(i_player1 == i_player2):
        return 0
    elif(i_player1 == "PEDRA" and i_player2 == "PAPEL"):
        return 2
    elif(i_player1 == "PEDRA" and i_player2 == "TESOURA"):
        return 1
    elif(i_player1 == "PAPEL" and i_player2 == "PEDRA"):
        return 1
    elif(i_player1 == "PAPEL" and i_player2 == "TESOURA"):
        return 2
    elif(i_player1 == "TESOURA" and i_player2 == "PEDRA"):
        return 2
    elif(i_player1 == "TESOURA" and i_player2 == "PAPEL"):
        return 1

winnable = 2
iterator = 0
maq_wins = 0
use_wins = 0
empates = 0

while iterator < winnable:
    
    random_choice = random.choice(rps_options)
    random_choice = random_choice.upper()

    option_user = input("\n Escolha papel, pedra ou tesoura:")    
    option_user = option_user.upper()
    
    while (option_user != "PEDRA" and option_user != "PAPEL" and option_user != "TESOURA"):
        option_user = input("Escolha somente entre as opções de papel, pedra ou tesoura:")
        option_user = option_user.upper()
    
    choicesPedraPapelTesoura(random_choice,option_user)
    resultado = pedraPapelTesoura(random_choice,option_user)
    
    if(resultado == 0):
        print("Empate")
        empates += 1
    if(resultado == 1):
        print("Vitória máquina")
        maq_wins += 1
     
    if(resultado == 2):
        print("Vitória usuário \n")
        use_wins += 1
    
    iterator = max(maq_wins,use_wins)
   
    
print("Placar \n Usuário:", use_wins, "\n  Máquina:", maq_wins, "\n Empates:", empates)
if(maq_wins > use_wins):
    print("Máquina venceu")
else:
    print("Usuário venceu")
```
