```python
#Criação de novas colunas para divisão de datas
#Existe a possibilidade de códigos mais eficientes para tratar datas, não necessariamente esta é a melhor solução

days = {name: i for i, name in enumerate(calendar.day_name)}

def treatDate(dt_time, cod_dt):
    date = (datetime.datetime.strptime(dt_time, "%m/%d/%Y %H:%M"))
    if(cod_dt == 1):
        return date.day
    elif(cod_dt == 2):
        return date.month
    elif(cod_dt == 3):
        return date.year
    elif(cod_dt == 4):
        return date.hour
    elif(cod_dt == 5):
        searched_key = date.weekday()
        for k, v in days.items():    
            if v == searched_key:
                return k

trips['start_day'] = trips['start_date'].apply(treatDate, cod_dt=1)
#trips['end_day'] = trips['end_date'].apply(treatDate, cod_dt=1)
trips['start_month'] = trips['start_date'].apply(treatDate, cod_dt=2)
#trips['end_month'] = trips['end_date'].apply(treatDate, cod_dt=2)
trips['start_year'] = trips['start_date'].apply(treatDate, cod_dt=3)
#trips['end_year'] = trips['end_date'].apply(treatDate, cod_dt=3)
trips['start_hour'] = trips['start_date'].apply(treatDate, cod_dt=4)
#trips['end_hour'] = trips['end_date'].apply(treatDate, cod_dt=4)
trips['start_dayofweek'] = trips['start_date'].apply(treatDate, cod_dt=5)
#trips['start_dayofweek'] = trips['start_date'].apply(treatDate, cod_dt=5)

trips['start_date'] = (trips['start_date'].apply(lambda x : pd.to_datetime(str(x))))

trips.head()
```

A variável lost_trips foi criada baseada no fato de que algumas bicicletas são locadas por muito tempo. 

Assumption 1: A média de viagens dentro do período de 1 hora é de 10 minutos.

Assumption 2 (exemplo): Há 100 corridas com duração de 2 horas cada. 
    
Problema: As bicicletas podem ficar ocupadas por 2 horas, causando indisponibilidade para corridas dentro da média considerada normal.     

1 corrida de 1 hora pode ser dividida em 6 corridas de 10 minutos. 
Logo 100 corridas de 2 horas podem ser divididas em 1200 corridas de 10 minutos. 

Isto é interessante quando se pensa no preço e estratégia dos serviços de bike sharing. 
Se a intenção é manter a disponibilidade de bicicletas nas estações de maior movimento, logo é não é ideal que viagens sejam muito longas. 
O preço pode ser modificado numa escala para que viagens mais longas paguem um preço mais caro, caso esta seja a estratégia. 


```python
pd.options.mode.chained_assignment = None
media_viagens = trips14[trips14['duration']<=60]['duration'].mean()

def calculateCostinRides(x):
    
    x = round(x/media_viagens,0)
    return int(x)
    
trips['lost_trips'] = trips[trips['duration']>60]['duration'].apply(calculateCostinRides);
```

Há muitas premissas a serem seguidas quando avaliamos correlaçao com precipitação. Elas não foram aplicadas aqui.

A intenção era calcular uma correlação da chuva em um ponto de São Francisco, com o número de corridas, que podem partir de qualquer estação dentro de São Francisco. O ponto de cada estação não necessariamente é o ponto que a chuva foi medida e este já é um problema. 

Mas para prática, vamos considerar a correlação sem premissas muito complicadas.
