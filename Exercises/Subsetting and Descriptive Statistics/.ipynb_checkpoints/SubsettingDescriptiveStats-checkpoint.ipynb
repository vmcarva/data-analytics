#### Show a visual summary of the data using boxplot. What Are the First and Third Quartiles? Comment your results.
##### * Hint : Quantiles vs Quartiles*
##### - In Probability and Statistics, quantiles are cut points dividing the range of a probability distribution into continuous intervals with equal probabilities. When division is into four parts the values of the variate corresponding to 25%, 50% and 75% of the total distribution are called quartiles.

```python
# draw boxplot here
bp_employee_salary = employee.boxplot('Salary')
employee['Salary'].describe()

#Q1 = 25% = 35
#Q3 = 75% = 60

#O boxplot em palavras mais simples desenha a função describe. A mediana se encontra no 50%, Q1 é o valor 25% e Q3 é o valor 75%
#Neste boxplot, os salários estão mais distribuídos no Q1 do que no Q3, indicando que mais pessoas ganham menos. 
```

```python
bp_employee_years = employee.boxplot('Years')
employee['Years'].describe()

#Q1 = 25% = 2
#Q3 = 75% = 7

#O boxplot em palavras mais simples desenha a função describe. A mediana se encontra no 50%, Q1 é o valor 25% e Q3 é o valor 75%
#Neste boxplot, os anos na empresa estão mais distribuídos no Q3 do que no Q1, indicando que a maior parte das pessoas tem 
# mais anos pela empresa. 
```

```python
# print first quartile here
print(employee['Years'].quantile([0.25]))
employee['Salary'].quantile([0.25])
```

```python
# print third quartile here
print(employee['Years'].quantile([0.75]))
employee['Salary'].quantile([0.75])
```
