```python
doors = { "door_a": {"name": "door a"},
          "door_b": {"name": "door b"},
          "door_c": {"name": "door c"},
          "door_d": {"name": "door d"}
        }

keys = {  "key_a": { "name": "key for door a", "target": "door_a"},
          "key_b": { "name": "key for door b", "target": "door_b"},
          "key_c": { "name": "key for door c", "target": "door_c"},
          "key_d": { "name": "key for door d", "target": "door_d"}
        } 

rooms = {"game_room":  { "name": "game room"},
         "bedroom_1":  { "name": "bedroom 1"},
         "bedroom_2":  { "name": "bedroom 2"},
         "living_room":  { "name": "living room"},
         "outside":  { "name": "outside"},
        }
         
furniture = ["couch", "piano", "queen bed", "double bed", "dresser", "dining table"]

object_relations = {
    "game room": [furniture[0], furniture[1], doors["door_a"]["name"]],
    "bedroom 1": [furniture[2], doors["door_a"]["name"], doors["door_b"]["name"], doors["door_c"]["name"]],
    "bedroom 2": [furniture[3], furniture[4], doors["door_b"]["name"]],
    "living room": [furniture[5], doors["door_c"]["name"], doors["door_d"]["name"]],
    "piano": keys["key_a"],
    "queen bed": keys["key_b"],
    "double bed": keys["key_c"],
    "dresser": keys["key_d"],
    "door a": [ rooms["bedroom_1"], rooms["game_room"] ],
    "door b": [ rooms["bedroom_2"], rooms["bedroom_1"] ],
    "door c": [ rooms["bedroom_1"], rooms["living_room"] ],
    "door d": [ rooms["living_room"], rooms["outside"] ]
}


INIT_GAME_STATE = {
    "current_room": rooms["game_room"],
    "keys_collected": [],
    "target_room": rooms["outside"]
}

game_state = INIT_GAME_STATE.copy()

def linebreak():
    """
    Print a line break
    """
    print("\n")

def start_game():
    """
    Start the game
    """
    print("You wake up on a couch and find yourself in a strange house with no windows which you have never been to before. You don't remember why you are here and what had happened before. You feel some unknown danger is approaching and you must get out of the house, NOW!")
    play_room(game_state["current_room"])


def play_room(room):
    """
    Play a room. First check if the room being played is the target room.
    If it is, the game will end with success. Otherwise, let player either 
    explore (list all items in this room) or examine an item found here.
    """
    game_state["current_room"] = room
    if(game_state["current_room"] == game_state["target_room"]):
        print("Congrats! You escaped the room!")
        return None
    else:
        linebreak()
        print("You are in " + room["name"])
        linebreak()
        intended_action = input("What would you like to do? Type 'explore' or 'examine'? ").strip()
        intended_action = intended_action.lower()
        if intended_action == "explore":
            explore_room(room)
        elif intended_action == "examine":
            examine_item(input("What would you like to examine? ").strip(), room)
        elif intended_action == "exit":
            print("Você desistiu do jogo")
            return None
        else:
            print("Not sure what you mean. Type 'explore' or 'examine'. ")
            play_room(room)
        
        if(game_state["current_room"] != game_state["target_room"]):
            linebreak()

def explore_room(room):
    """
    Explore a room. List all items belonging to this room.
    """
    items = ", ".join(object_relations[room["name"]])        
    print("This is " + room["name"] + ". Here you find the following items:", items)
    play_room(room)
    
def examine_item(item_name, room):
    """
    Examine an item which can be a door or furniture.
    First make sure the intended item belongs to the current room.
    Then check if the item is a door. Tell player if key hasn't been 
    collected yet. Otherwise ask player if they want to go to the next
    room. If the item is not a door, then check if it contains keys.
    Collect the key if found and update the game state. At the end,
    play either the current or the next room depending on the game state
    to keep playing.
    """

    item_name = item_name.lower()
    current_room = game_state["current_room"]
    next_room = ""
    output = None
    have_key = False
    
 
    for item in object_relations[room["name"]]:
        if(item == item_name):
            output = "You examine " + item_name + ". "
            if(item_name[:4] in "door"):    
#               if(item in game_state["keys_collected"][0]["name"]):
                if(game_state["keys_collected"]):
                    if(hasKey(game_state["keys_collected"],item)):
                        have_key = True
                if(have_key):
                    output += "You unlock it with a key you have."
                    next_room = get_next_room_of_door(item, current_room)
                else:
                    output += "It is locked and you don't have the key."
            else:

                if(item in object_relations and len(object_relations[item])>0):
                    item_found = object_relations[item]
                    object_relations.pop(item)
                    game_state["keys_collected"].append(item_found)
                    output += "You found the " + item_found["name"] + "."
                else:
                    output += "There isn't anything interesting about it."
            print(output)
            break

    if(output is None):
        print("The item you requested is not found in the current room.")
    
    if(room == game_state["target_room"]):
        print("Parabéns, você cumpriu a missão do escape game!")
        return None
    
    proximaSalaTeste = None
    if(next_room and proximaSalaTeste == None):
        proximaSalaTeste = input("Are you sure you want to go to the next room? Enter 'yes' or 'no' ").strip()
        proximaSalaTeste = proximaSalaTeste.lower()
        if(next_room and proximaSalaTeste == 'yes'):
            play_room(next_room)
        elif(next_room and proximaSalaTeste == 'no'):
            play_room(current_room)    
        else:
            print("Not sure what you mean. Type 'yes' or 'no'. ")
            play_room(current_room)
    else:
        play_room(current_room)

def hasKey(lst, item):
    for i in range(0,len(lst)):
        if item in lst[i]["name"]:
            return True
    else:
        return False
        

def get_next_room_of_door(door, current_room):
    """
    From object_relations, find the two rooms connected to the given door.
    Return the room that is not the current_room.
    """
    next_entrance = ""
    connected_rooms = object_relations[door]
    for room in range(0,len(connected_rooms)):
        if(current_room["name"] != connected_rooms[room]["name"]):
            next_entrance = connected_rooms[room]
    return next_entrance

start_game()

# voltar para sala anterior

```
